```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
# draw 2d gaussian distribution
x_var, y_var = np.random.multivariate_normal([0, 0], np.identity(2), 1000).T
```

```python
# plot
plt.figure()
plt.hist2d(x_var, y_var, bins=50, range=np.array([(-10, 10), (-10, 10)]))
plt.show()
```

```python
zx_var = 2 * x_var + 3
zy_var = 2 * y_var + 3
```

```python
# plot
plt.figure()
plt.hist2d(zx_var, zy_var, bins=50, range=np.array([(-10, 10), (-10, 10)]))
# plt.axis('equal')
plt.show()
```

```python
# plot
plt.figure()
plt.hist(x_var, bins=50, range=(-10, 10))
plt.show()
```

```python
# plot
plt.figure()
plt.hist(zx_var, bins=50, range=(-10, 10))
plt.show()
```
